```python
#1
def sum_even(*args):
    answer = 0
    if len(args) == 0:
        return -1
    else:     
        for data in args:
            if data%2==0:
                answer += data
        return answer
```

```python
sum_even(2,55,6,7,4)
```

```python
#3
def make_list(*a, type=10):
    answer = []
    for i in a:
        answer.append(i*type)
    return answer
```

```python
print(make_list(1,2,3,4,5))
print(make_list(1,2,3,4,5, type=3))
print(make_list(10, 20, 30, 40, 50, 60, type=2))
print(make_list(1,2,3, type=100))
print(make_list(1,1,1,1,1,1,1,type=0))
```

```python
#4
answer = []
def update_list(nums, inplace=False):
    if inplace:
        for i in range(len(nums)):
            nums[i] = nums[i]**2
    else:
        num_list = []
        for data in nums:
            num_list.append(data**2)
        return num_list
```

```python
mynums = [ 3, 7, 2, 6, 10]
print(update_list(mynums, False))
print(mynums)
print(update_list(mynums, True))
print(mynums)
```

```python
#5
dict = {}
def mydic(**d):
    if len(d) == 0:
        return d
    else:
        for key, value in d.items():
            dict["my " + key] = value
        return dict
```

```python
mydic(b=2, c=3, d=4, e=5, f=6)
```

```python
#6
def make_dict(*nums):
    dict = {}
    even = 1
    odd = 1
    for data in nums:
        if data % 2 == 0:
            dict[f"even{even}"] = data
            even+=1
        elif data % 2 != 0:
            dict[f"odd{odd}"] = data
            odd+=1
        else:
            pass
    return dict
```

```python
print(make_dict(1,2,3,4,5))
print(make_dict(10,23,45,5))
print(make_dict())
```

```python
#7
def myprint(*d, deco="**", sep=","):
    if len(d) == 0:
        print("Hello Python")
        return
    else:
        result = ""
        for data in range(len(d)):
            result += str(d[data])
            if data != len(d)-1:
                result += sep
        return deco+result+deco
```

```python
print(myprint(10, 20, 30, deco="@", sep="-"))
print(myprint("python", "javascript", "R", deco="$"))
print(myprint("가", "나", "다") )
print(myprint(100) )
print(myprint(True, 111, False, "abc", deco="&", sep="") )
```
