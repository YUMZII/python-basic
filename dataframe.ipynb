```python
#1
import pandas as pd
data = {
 'name':['둘리', '또치', '도우너', '희동이'],
 'kor':[90, 80, 70, 70],
 'eng':[99, 98, 97, 46],
 'mat':[90, 70, 70, 60],
}
df = pd.DataFrame(data)
print(df)
```

```python
df['class']=['1반','2반','1반','2반']
print(df)
```

```python
#2
df.loc[len(df)]={'name':'마이콜','kor':80, 'eng':80, 'mat':80, 'class':'1반'}
print(df)
```

```python
#3
df.set_index('name',inplace=True)
print(df)
```

```python
df.rename(columns={'kor':'국어','eng':'영어','mat':'수학', 'class':'반번호'}, inplace=True)
print(df)
```

```python
#4
df.loc['마이콜','국어']=100
df.loc['마이콜','영어']=100
df.loc['마이콜','수학']=100
df.loc['희동이','영어']=90
print(df)
```

```python
#5
df = df.reset_index() #인덱스 초기화
print(df)
df.rename(columns={'name':'성명'}, inplace=True)
print(df)
```

```python
#6
df1=df.sort_values(by='국어', ascending=False)
print(df1)
df2=df.sort_values(by='영어')
print(df2)
```

```python
#7
df["총점"] = df.국어 + df.영어 + df.수학 #sum 사용할 수 있음
df = df.sort_values(by='총점', ascending=False)
print(df)
```

```python
#8 axis 행 = 0 열 = 1
df3 = df.copy()
df3.drop('반번호', axis=1, inplace=True)
print(df3)
```

```python
#8
df4 = df.copy()
df4.drop(4, inplace=True)
print(df4)
print(df)
```
