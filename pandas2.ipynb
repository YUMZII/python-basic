```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
df = pd.read_csv("./data/cctv_seoul.csv")
display(df)
```

```python
#2
df.set_index('구별', inplace=True)
display(df)
```

```python
#3
df.sort_values(by='CCTV수',ascending=False)
```

```python
#4 ['CCTV수'] 출력하고 싶은 것만
ndf = df.sort_values(by='CCTV수')['CCTV수']
display(ndf)
```

```python
length = len(ndf); length
```

```python
#4
#한글 없다고 깨질 때
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

mycmap2 = plt.get_cmap("Spectral", length)#리턴값이 함수
mycolor2 = [mycmap2(i) for i in range(length)]#mycmap2(i) 출력해야되는 수 만큼
ndf.plot(kind="barh", color=mycolor2)
plt.title('구별 CCTV 현황')
```

```python
#5
df['CCTV비율'] = (df['CCTV수']/df['인구수'] * 100).round(6)
df['노인비율'] = (df['CCTV수']/df['인구수'] * 100).round(6)
display(df)
```

```python
#7
mycmap1 = plt.get_cmap("PRGn", length)
mycolor1 = [mycmap1(i) for i in range(length)]
```

```python
df['CCTV비율'].plot(kind="bar", color=mycolor1)
plt.title('인구수 대비 CCTV 비율')
```
