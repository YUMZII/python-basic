```python
#2
df_1 = pd.read_csv("./data/score_noheader.csv")
display(df_1)
print("컬럼명: ", df_1.columns)
print("인덱스: ",df_1.index)
print("컬럼명 추가 ----------")
```

```python
df_1.rename(columns={"둘리":"name", "90" :"kor", "90.1": "eng", "90.2": "mat"}, inplace=True)
print("컬럼명: ", df_1.columns)
display(df_1)
```

```python
print("total과 avg추가----------")
df_1["total"] = df_1[['kor', 'eng', 'mat']].sum(axis=1)
df_1["avg"] = df_1[['kor', 'eng', 'mat']].mean(axis=1).round(2)

display(df_1)
```

```python
#3
df = pd.read_csv("./data/score_header.csv")
display(df)
df.drop(0, inplace=True)
df.drop(1, inplace=True)
df.drop(2, inplace=True)
display(df)
```

```python
df.rename(columns ={'성적파일입니다 ':'name','Unnamed: 1':'kor','Unnamed: 2': 'eng','Unnamed: 3':'mat'}, inplace=True)
display(df)
#문자열을 숫자로 변환
df[['kor','eng','mat']] = df[['kor','eng','mat']].apply(pd.to_numeric)

df["total"] = df[['kor', 'eng', 'mat']].sum(axis=1)
df["avg"] = df[['kor', 'eng', 'mat']].mean(axis=1).round(2)
display(df)
```

```python
for i in range(3):
    df.to_csv(
        "output/score_result.csv",
        mode='a',
        header=False,
        index=False
    )
```
