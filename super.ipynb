```python
#2
class Member:
    def __init__(self, name, account, passwd, birth_year):
        self.name = name
        self.account = account
        self.passwd = passwd
        self.birth_year = birth_year
    def __str__(self):
        return f'{self.name}({self.account},{self.passwd},{self.birth_year})'
```

```python
members = [Member("조윤지", "윤지",'4444','2000'), 
           Member("조윤지", "윤지",'4444','2000'),
           Member("조윤지", "윤지",'4444','2000'),]
for i, members in enumerate(members, 1):
    print(f'회원{i}:{members}')
```

```python
#3
while True:
    score = input("숫자를 입력하세요: ")
    try:
        score = int(score)
        answer = 0
        for i in range(1, score + 1):
            answer += i
        print(f'1부터 {score}까지의 합은 {answer}입니다.')
        break
    except:
        print("숫자만 입력해 주세요~~~ ")
```

```python
#4
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
        
    def getBookInfo(self):
        return f"{self.title}\n{self.author}\n{self.price}"
        
    def __str__(self):
        return "Book 클래스로 생성된 객체"
```

```python
books = [Book("자바의 정석", "저자", 19000),
    Book("파이썬 완벽 가이드", "저자", 25000),
    Book("자료구조", "저자", 30000)]

for i, book in enumerate(books, 1):
    print(f"[객체{i}:{book}]")
    print(book.getBookInfo())
    print("*"*50)
```

```python
#5
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def get_point(self):
        return self.price*0.05
    def __str__(self):
        return f"{self.name} 상품의 가격은 {self.price:,}원입니다."
```

```python
p1 = Product("둘리인형", 30000)
print(p1)
print(f"포인트 정보 : {p1.get_point()}")
p2 = Product("노트북",1000000)
print(p2)
print(f"포인트 정보 : {p2.get_point()}")
```

```python
#6
class Employee:
    def __init__(self, name, salary, lan):
        self.name = name
        self.salary = salary
        self.lan = lan
    def show_info(self):
        return f"이름은 {self.name}, 급여는 {self.salary:,}원입니다 "
class Developer(Employee):
    def __init__(self, name, salary, lan):
        super().__init__(name, salary, lan)
    def show_info(self):
        return super().show_info() + f"{self.lan}프로그래밍이 가능함"
```

```python
dev = Developer("둘리", 150000, "Java")
print(dev.show_info())
dev = Developer("또치", 120000, "Python")
print(dev.show_info())
dev = Developer("도우너", 110000, "JavaScript")
print(dev.show_info())
```
