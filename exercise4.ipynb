```python
listnum = [10, 5, 7, 21, 4, 8, 18]
print(listnum)
```

```python
count = 0

for i in listnum:
    if i % 2==0:
        count +=1
print(f"짝수의 갯수:{count}개")
```

```python
listnum.sort()
min1 = listnum[0]
print(f'최솟값: {min1}')
```

```python
max1 = listnum[0]
for num in listnum:
    if num > max1:
        max1 = num
print(f'최댓값: {max1}')
```

```python
def get_min_max(listnum2):
    min_v = min(listnum2)
    max_v = max(listnum2)
    print(f'생성된 리스트 : {listnum2}')
    print(f'최솟값은{min_v}이고 최댓값은 {max_v} 입니다.')
    return
```

```python
listnum2 = []
for i in range(10):
    listnum2.append(random.randint(1,30))
get_min_max(listnum2)
```
