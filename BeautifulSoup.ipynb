```python
import requests
from bs4 import BeautifulSoup

req = requests.get('http://unico2013.dothome.co.kr/crawling/exercise_bs.html')
html = req.content

bs = BeautifulSoup(html, 'html.parser')
hrefs =bs.select('a')
imgs = bs.select('img')
colors = bs.find('h2')
uls = bs.find('li', string='녹색')
foods = bs.find_all('h2')[1]
food_kind = bs.select('ol')
name = bs.select('table')
name_tr = bs.find('tr', class_='name')
targets = bs.find('td', id='target')

print("[<h1>태그의 컨텐츠]HTML의 링크 태그")
print("[텍스트 형식으로 내용을 가지고 있는 <a>태그의 컨텐트와 href속성값]")
for tag in hrefs: #텍스트가 있는 a태그만 출력
    if tag.text.strip():# 공백제거
        print(tag.text, ":", tag['href'])
print("[<img> 태그의 src 속성값]")  
for img in bs.select('img'):
    print(img.get('src'))
print("[첫 번째 <h2> 태그의 콘텐츠]", colors.text )
print("[<ul> 태그의 자식 태그들 중 style 속성의 값이 green으로 끝나는 태그의 콘텐츠]", uls.text)
print("[두 번째 <h2> 태그의 콘텐츠]", foods.text)
print("[<ol> 태그의 모든 자식 태그들의 콘텐츠]")
for tag2 in food_kind:
    print(tag2.text)
print("[<table> 태그의 모든 자손 태그들의 콘텐츠]")
for tag3 in name:
    print(tag3.text)
print("[name이라는 클래스 속성을 갖는 <tr> 태그의 콘텐츠]", name_tr.text)
print("[target이라는 아이디 속성을 갖는 <td> 태그의 콘텐츠]",targets.text )
```
