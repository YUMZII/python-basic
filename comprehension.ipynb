```python
#3 [결과값 for 변수 in 반복대상 if 조건]

def createList(*a, type=1):
    if len(a) == 0:
        return [i for i in range(1,10)]
    elif type == 1:
        return [i for i in a]
    elif type == 2:
        result =[i for i in a if not i%2]
        return result
    elif type == 3:
        result = [i for i in a if i%2]
        return result
    elif type == 4:
        result = [i for i in a if 10 < i]
        return result
```

```python
print(createList())
print(createList(66,44,77 , type=1))
print(createList(66,44,77 , type=2))
print(createList(66,44,77 , type=3))
print(createList(66,44,77 , type=4))
```

```python
#4
def mycompre_dict(**m):
    if len(m) == 0: 
        return m
    dict ={ "my "+ str(k): v for k,v in m.items() }
    return dict
```

```python
print(mycompre_dict(b=2, c=3, d=4, e=5, f=6))
print(mycompre_dict())
```

```python
#5
def make_comprelist(*nums, type = 10):
    make_list=[i * type for i in nums]
    return make_list
```

```python
print(make_comprelist(1,2,3,4,5))
print(make_comprelist(1,2,3,4,5, type=3))
print(make_comprelist(10, 20, 30, 40, 50, 60, type=2))
print(make_comprelist(1,2,3, type=100))
print(make_comprelist(1,1,1,1,1,1,1, type=0))
```
