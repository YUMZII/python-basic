```python
#5
blood = ['A형', 'B형', 'O형', 'AB형']
st = [34.2, 27.1, 26.7, 11.5]
sr = pd.Series(st, index=blood)

print("[시리즈 sr 출력]")
print(sr)
print("[시리즈 sr 에서 인덱싱과 슬라이싱하여 출력(숫자인덱스 사용)]")
print(st[0])
print(st[3])
print(st[1])
print(sr[1:])
print("[시리즈 sr 에서 인덱싱과 슬라이싱하여 출력(이름인덱스 사용)]")
print(sr['A형'])
print(sr['AB형'])
print(sr['B형'])
print(sr[1:])
```

```python
print("[시리즈 sr의 크기 출력]")
print(sr.size)
print("[시리즈 sr의 인덱스 출력]")
print(sr.index)
print("[시리즈 sr의 데이터 값 출력]")
print(sr.values)
```
