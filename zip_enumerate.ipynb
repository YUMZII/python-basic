```python
#1
def myemail_info(email):
    if '@'not in email:
        pass
    else:
        result = tuple(email.split("@"))
        return result
```

```python
print(myemail_info('1234@naver.com'))
print(myemail_info('1234naver.com'))
```

```python
#3
names = ["둘리", "또치", "도우너", "고길동"]
scores = [85, 92, 78, 96]

a=zip(names,scores)
for name, score in a:
    print(f"{name}:{score}점")
```

```python
#4
words = ["apple", "banana", "cherry"]
meanings = ["사과", "바나나", "체리"]

dic = zip(words, meanings)
b = {words:meanings for words, meanings in dic}
print(b)
```

```python
#5
fruits = ["배", "사과", "포도"]
counts = [5, 3, 2]
prices = [3000, 1500, 5000]

m = zip(fruits, counts, prices)
for fruits, counts, prices in m:
    print(f"{fruits}의 총가격은 {prices * counts}원입니다.")
```

```python
#6
animals = ["lion", "tiger", "bear", "rabbit", "fox"]
result=[]
for i, animals in enumerate(animals):
    if i%2 == 0:
        result.append(animals.upper())
    else:
        result.append(animals)
print(result)
```
