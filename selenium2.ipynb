```python
#1
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time

url = "http://gs25.gsretail.com/gscvs/ko/products/event-goods"
driver=webdriver.Chrome()
driver.get(url)
dictdata = []

#특정 요소 위치 이동
element = driver.find_element(By.PARTIAL_LINK_TEXT,'2+1 행사').click()
time.sleep(2)

# 1~10 페이지 정보
for page in range(1,115):
    #2+1상품명과 가격 추출
    products = driver.find_elements(By.CLASS_NAME, 'prod_list')
    for product in products:
        byClassNames = product.find_elements(By.CLASS_NAME, 'tit')
        byClassCosts = product.find_elements(By.CLASS_NAME, 'cost')
        for className,classCost in zip(byClassNames, byClassCosts):
            if not className.text.strip():
                continue
            if not classCost.text.strip():
                continue
            dictdata.append({"goodsname":className.text,"goodsprice":classCost.text})
    page_btn = driver.find_element(By.PARTIAL_LINK_TEXT,"다음 페이지로 이동").click()
    time.sleep(2)      
    print(f"{page} 페이지 스크래핑")
df = pd.DataFrame(dictdata)
print(f"{len(dictdata)}개 아이템 스크래핑 완료")
df.to_csv("output/gs25_twotoone.csv", index=False, encoding="utf-8")
```

```python
#2
from urllib import request
import json

api_key =''
num_disp = 1000

url = f'http://openapi.seoul.go.kr:8088/{api_key}/json/GetParkingInfo/1/{num_disp}/'
res = request.urlopen(url)
data = json.loads(res.read().decode("utf-8"))
rows = data["GetParkingInfo"]["row"]
parkList=[]
for itemList in rows:
    parking = itemList["PKLT_NM"]              
    time = itemList["NOW_PRK_VHCL_UPDT_TM"]  
    total_cnt = itemList["TPKCT"]              
    current_cnt = itemList["NOW_PRK_VHCL_CNT"]
    print(f"[주차장] 주차장명:{parking}|업데이트 시간:{time}|총 주차면 수:{total_cnt} | 현재 주차 차량 수:{current_cnt}")
```

```python
#2
from urllib import request
from bs4 import BeautifulSoup

api_key =''
num_disp = 1000

url = f'http://openapi.seoul.go.kr:8088/{api_key}/xml/GetParkingInfo/1/{num_disp}/'
res = request.urlopen(url)
xml = res.read()
soup = BeautifulSoup(xml, "xml")

for itemList in soup.find_all("row"):
    parking = itemList.find("PKLT_NM").text              
    time = itemList.find("NOW_PRK_VHCL_UPDT_TM").text  
    total_cnt = itemList.find("TPKCT").text              
    current_cnt = itemList.find("NOW_PRK_VHCL_CNT").text
    print(f"[주차장] 주차장명:{parking}|업데이트 시간:{time}|총 주차면 수:{total_cnt} | 현재 주차 차량 수:{current_cnt}")
```
